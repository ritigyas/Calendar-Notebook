```python
# Importing tkinter module
from tkinter import *
import calendar


# Function to show calendar of the given year
def showCalender():
    gui = Tk()  
    gui.config(background='blue')
    gui.title("Calendar for the year")
    gui.geometry("550x600")

    year = int(year_field.get())  # Get the entered year
    gui_content = calendar.calendar(year)

    calYear = Label(gui, text=gui_content, font="Consolas 10 bold")  # 'consolas 10 bold' is used for alignment of text such that the characters are equally spaced.
    calYear.grid(row=5, column=1, padx=20)  # padx means padding in x(horizontal) direction. It add space to the left and right of widget. 
    #Padding is the space between the content and the border of an element. 

    gui.mainloop()  
    #tkinter applications are event-driven, meaning they wait for user interactions (button clicks, text input, etc.). with this loop , application will not get open. 
    #mainloop() enters an infinite loop, continuously running the program until the user closes the window.

                    

# Driver Code
if __name__ == '__main__':                        
    new = Tk()
    new.config(background='grey')
    new.title("Calendar")
    new.geometry("250x140")

    cal = Label(new, text="Calendar", bg='grey', font=("times", 28, "bold" ) , fg = 'red')
    year = Label(new, text="Enter year", bg='dark grey')

    year_field = Entry(new)
    button = Button(new, text='Show Calendar', fg='Black', bg='pink', command=showCalender)
    exit_button = Button(new, text="Exit", command=new.destroy)  # Use destroy() instead of quit()           #quit()	- Stops the Tkinter mainloop(), but the window might not close completely.
                                                                                                             #destroy()	- Closes the window and frees up resources.
                                                                                                             
    # Putting widgets in position
    cal.grid(row=1, column=1)
    year.grid(row=2, column=1)
    year_field.grid(row=3, column=1)
    button.grid(row=4, column=1)

   
    exit_button.grid(row=6, column=1)

    new.mainloop()  # Start the main GUI loop


    #Every Python script has a special built-in variable called __name__.
    #When a Python file is run directly, __name__ is set to "__main__".
    #When a Python file is imported into another script, __name__ is set to the name of that script (module name).
```

```python
import calendar 
from tkinter import * 


def showcalendar() :

    gui = Tk()
    gui.config(background = 'green')
    gui.geometry("550x600")
    gui.title("Show a calendar")
    
    year = int(entry_field.get())
    gui_content = calendar.calendar(year)
    showing_years = Label(gui , text = gui_content , bg = 'pink' , font = "Consolas 10 bold" )  
    showing_years.grid(row = 5 , column =1  , padx =20)
    gui.mainloop()

if __name__ == '__main__' : 
    new = Tk()
    new.config(background = 'grey')
    new.geometry("250x140")
    new.title('Calendar')

    cal = Label(new , text = 'Calendar' , font = ('Times' , 30 , "bold"))
    enter_year = Label(new , text = 'Enter year' , fg = 'black')
    entry_field = Entry(new)

    showcal_button = Button(new , text = 'Show Calendar' , bg = 'pink' , fg = 'Blue' , command = showcalendar)
    exit_button = Button(new , text = "Exit" , bg = 'white' , fg = 'black' , command = new.destroy)

    cal.grid(row = 1 , column = 1)
    enter_year.grid(row = 2 , column = 1)
    entry_field.grid(row = 3 , column = 1)
    showcal_button.grid(row = 4 , column = 1)
    exit_button.grid(row = 6 , column = 1)

    new.mainloop() 
    
```

```python
from tkinter import *


window = Tk()
window.title('welcome to edureka')
window.geometry('350x200')

def left_click(event) :
    Label(window , text = 'left clicked').pack()

def middle_click(event) :
    Label(window , text = 'middle clicked').pack()

def right_click(event) :
    Label(window , text = 'right clicked').pack()



window.bind("<Button-1>" , left_click)        # given names to the button
window.bind("<Button-2>" , middle_click)
window.bind("<Button-3>" , right_click)

window.mainloop()
```
